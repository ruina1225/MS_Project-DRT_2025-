```python
# whisper 음성인식 실행
import whisper

# 모델 불러오기 (tiny, base, small, medium, large 중 선택 가능)
model = whisper.load_model("base")

# 오디오 파일 이름 입력 (예: audio.mp3)
result = model.transcribe("test.m4a")

# 결과 출력
print(result["text"])
```

```python
# 모델 불러오기 (tiny, base, small, medium, large 중 선택 가능)
model = whisper.load_model("medium")

# 비디오 파일 이름 입력 (예: audio.mp3)
result = model.transcribe("test video1.mp4")

# 결과 출력
print(result["text"])
```
